```python
# Rename column is_smoking to smoking status..
df = df.rename(columns={'is_smoking': 'smoking status'})
```

```python
# To check the data again, is there any missing values or not, so there is no null values here...
df
```

```python
# mean of column 'BMI'
df['BMI'].mean()
```

```python
# mean of column 'glucose'
df['glucose'].mean()
```

```python
# drop the values..
df.dropna(axis=0, inplace=True)
df.head()
```

```python
# fill the null values with mean..
df['education']= df['education'].fillna(df['education'].mean())
df['BPMeds']= df['BPMeds'].fillna(df['BPMeds'].mean())
df['BMI']= df['BMI'].fillna(df['BMI'].mean())
df['totChol']= df['totChol'].fillna(df['totChol'].mean())
df['glucose']= df['glucose'].fillna(df['glucose'].mean())
```

```python
df_cigs = df[['sex','cigsPerDay']]
```

```python
grouped_cigs = df_cigs.groupby(['sex'],as_index=False).mean()
```

```python
# male are smoking more cigrettes per day (13.7), however, females are smoking lesser than males (5.4 per day)
grouped_cigs.head()
```

```python
# make a pie chart of cigsperday with the value hue value 'sex', which is showing males percentage(71.7%)
# are more than female percentile(28.3%), cigrettes consumes by male more than female...
plt.pie(grouped_cigs['cigsPerDay'], labels= grouped_cigs['sex'], autopct='%1.1f%%', shadow = True)
plt.legend(title = 'CIGS PER DAY BY SEX')
plt.show()
```
